# Friedman Data

## Purpose
1. Needed to test hypothesis that Isolation Forests can predict anomalies in a dataset. 

## Visualization of Friedman Data

```python
# Ex 1: Visualizing Friedman dataset in 2D
from sklearn.datasets import make_friedman1
import matplotlib.pyplot as plt

X, y = make_friedman1(n_samples=100, n_features=10, noise=0.0, random_state=None)

plt.plot(X, y, marker='o', color='red', linestyle='None')
plt.show()
```

```python
# Ex 2: Visualizing Friedman dataset output in 2D with noise of 150
import matplotlib.pyplot as plt

"""
The y values fall in the range of -600 to 600, for this example. If we already have a trained model and we 
predict an output outside of this range, Then that instance will be detected as an anomaly if we train the iForest 
model correctly.
"""
X, y = make_friedman1(n_samples=100000, n_features=5, noise=150, random_state=1)

plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(y)
plt.show()
```

## Using Isolation Forests on the Friedman Data

The plan of action is to identify anomalies based on the output. This is because:
1. The input variables are uniform, so trying to detect anomalies based on those values will be ineffective
2. Identifying anomalies based on the output (aka y) would be helpful.
